#### importing numpy library

```python
import numpy as np
```

```python
#creating an array with numpy
arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

#### The array object in numpy is called ndarray

```python
print(type(arr))
```

#### we can pass a list, tuple or any array-like object into the array() method, and it will be converted into an ndarray

```python
mytuple = ("apple", "banana", "mango") 

arr2 = np.array(mytuple)

print(arr2)
```

#### ndarrays can have many dimensions

```python
a = np.array(42)
b = np.array([1, 2, 3, 4, 5])
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)
```

#### for higher dimensional arrays we can use __ndim__ argument to specify the number of dimensions

```python
e = np.array([1, 2, 3, 4, 5], ndmin=5)

print(e)
print('number of dimensions: ', e.ndim)
```

#### ndarrays are 0 indexed

```python
print(arr[0]) #1st element of the array
print(arr[1]) #2nd element of the array
print(arr[2] + arr[3])
```
